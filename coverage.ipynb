```python
def mean_area_cls(ma):
    if ma >= 5000:
        return ">5000"
    elif ma >=3000:
        return "3000-5000"
    elif ma >=1500:
        return "1500-3000"
    else:
        return "<1000"

yd_g['ma_cls'] = yd_g.MEAN_AREA.apply(mean_area_cls)
```

```python
fig = plt.figure(figsize=(10,6), dpi=600)

#ax = sns.histplot(data=yd_g, x="coverage",log_scale=True,kde=True,)
ax = sns.histplot(data=yd_g, x="coverage",hue='ma_cls',bins=60,element="step",
                  fill=False,
                  log_scale=(False,True)
                  #,color='grey'
                 )
ax.set_xlabel('Yardang cover (%)', fontdict={'size': 10})
```
